Interactive 3D plot of the movement from the cnc test rig

```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio


df = pd.read_csv("C:/Users/maxjj/OneDrive/Dokumenter/Skole/bachelor/csvfiler/stvg_friday_3-x300y300.csv")
df.columns = ["pos_x", "pos_y", "pos_z", "vel_x", "vel_y", "vel_z"]

print(df.columns)
data= df.copy(deep=True)



fig = px.line_3d(data, 
                    x=data['pos_x'], 
                    y=data['pos_y'], 
                    z=data['pos_z'],
                    # color = data["z"], 
                    # color_continuous_scale='viridis', 
                    # size_max=2, opacity=0.7,
                    # width=230,
                    # animation_frame=""
                    )


fig.update_layout(
    scene = dict(
        xaxis = dict(nticks=4, range=[0,1],),
        yaxis = dict(nticks=4, range=[0,1],),
        zaxis = dict(nticks=4, range=[-.3,0],),),
    
    width=700,
    margin=dict(r=20, l=10, b=10, t=10))

fig.update_traces(
    marker=dict(
        color='lightblue',
        size=10),
    line=dict(
        # color='purple',
        width=5
    )
)

fig.update_layout(
    autosize=False,
    width = 1000,
    height = 750,
)
fig.write_image("csvfiler/trace1.png", scale = 2)

fig.show()
```

matplotlib 3D plot of the movement from the test rig

```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
import matplotlib.pyplot as plt


df = pd.read_csv("C:/Users/maxjj/OneDrive/Dokumenter/Skole/bachelor/csvfiler/stvg_friday_3-x300y300.csv")
df.columns = ["pos_x", "pos_y", "pos_z", "vel_x", "vel_y", "vel_z"]

print(df.columns)
data= df.copy(deep=True)

ax = plt.figure().add_subplot(projection='3d')

z = data['pos_z']
x = data['pos_x']
y = data["pos_y"]

ax.plot(x, y, z, label='parametric curve')
ax.legend()

ax.view_init(elev=10,azim=0,roll=0)
plt.show()
```

matplotlib 3D plot of the movement from the simulation

```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
import matplotlib.pyplot as plt


df = pd.read_csv("csvfiler/gz_with_current_xy.csv")
# df.columns = ["pos_x", "pos_y", "pos_z", "vel_x", "vel_y", "vel_z"]

# print(df.columns)
df = df.drop(df.columns[7:], axis=1)
df = df.drop(df.columns[[2,3]], axis=1)

new_columns = ["sekunder","nanosekunder","pos_x","pos_y","pos_z"]
df.columns = new_columns
df["nanosekunder"] = df["nanosekunder"]*(10**-9)
df["time"] = df["sekunder"]+df["nanosekunder"]
df = df[df["pos_z"]>.026]


data= df.copy(deep=True)

ax = plt.figure().add_subplot(projection='3d')

z = data['pos_z']
x = data['pos_x']
y = data["pos_y"]

ax.plot(x, y, z, label='Movement Path')
ax.scatter(0.0,0.0,0.02,c="red",marker="x")
ax.legend()

ax.view_init(elev=0,azim=90,roll=0)
plt.show()
```
